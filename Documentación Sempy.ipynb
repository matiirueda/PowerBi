### Funciones de Lista

###### El propÃ³sito de utilizar las funciones de lista de Sempy es crear un catÃ¡logo exhaustivo de los objetos dentro de Power BI, proporcionando una visiÃ³n integral de los componentes y la estructura de los modelos semÃ¡nticos. Estas funciones permiten enumerar y documentar detalladamente capacidades, columnas, conjuntos de datos, expresiones, jerarquÃ­as, Ã­tems, medidas, particiones, solicitudes de actualizaciÃ³n, violaciones de relaciÃ³n, relaciones, informes, tablas y traducciones, asÃ­ como espacios de trabajo. Al utilizar estas funciones, podrÃ¡s obtener un panorama completo de los recursos disponibles, facilitando la gestiÃ³n, anÃ¡lisis y optimizaciÃ³n de los modelos y reportes de Power BI.

- list_capacities(): Devuelve una lista de capacidades disponibles.
- list_columns(): Enumera todas las columnas en un conjunto de datos.
- list_datasets(): Proporciona una lista de todos los conjuntos de datos.
- list_expressions(): Enumera expresiones, como fÃ³rmulas DAX en modelos.
- list_hierarchies(): Devuelve una lista de jerarquÃ­as utilizadas en los modelos. 
- list_items(): Enumera todos los Ã­tems, como informes y conjuntos de datos, en un espacio de trabajo.
- list_measures(): Lista todas las medidas en un conjunto de datos.
- list_partitions(): Proporciona informaciÃ³n sobre las particiones en un conjunto de datos.
- list_refresh_requests(): Enumera las solicitudes de actualizaciÃ³n para los conjuntos de datos.
- list_relationship_violations(): Muestra las violaciones de relaciÃ³n en los modelos.
- list_relationships(): Enumera las relaciones entre tablas en un modelo.
- list_reports(): Devuelve una lista de informes.
- list_tables(): Enumera las tablas en un conjunto de datos.
- list_translations(): Proporciona las traducciones disponibles para los elementos en un modelo.
- list_workspaces(): Enumera los espacios de trabajo disponibles.
