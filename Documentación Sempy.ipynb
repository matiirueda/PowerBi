### Funciones de Lista

###### El propósito de utilizar las funciones de lista de Sempy es crear un catálogo exhaustivo de los objetos dentro de Power BI, proporcionando una visión integral de los componentes y la estructura de los modelos semánticos. Estas funciones permiten enumerar y documentar detalladamente capacidades, columnas, conjuntos de datos, expresiones, jerarquías, ítems, medidas, particiones, solicitudes de actualización, violaciones de relación, relaciones, informes, tablas y traducciones, así como espacios de trabajo. Al utilizar estas funciones, podrás obtener un panorama completo de los recursos disponibles, facilitando la gestión, análisis y optimización de los modelos y reportes de Power BI.

- list_capacities(): Devuelve una lista de capacidades disponibles.
- list_columns(): Enumera todas las columnas en un conjunto de datos.
- list_datasets(): Proporciona una lista de todos los conjuntos de datos.
- list_expressions(): Enumera expresiones, como fórmulas DAX en modelos.
- list_hierarchies(): Devuelve una lista de jerarquías utilizadas en los modelos. 
- list_items(): Enumera todos los ítems, como informes y conjuntos de datos, en un espacio de trabajo.
- list_measures(): Lista todas las medidas en un conjunto de datos.
- list_partitions(): Proporciona información sobre las particiones en un conjunto de datos.
- list_refresh_requests(): Enumera las solicitudes de actualización para los conjuntos de datos.
- list_relationship_violations(): Muestra las violaciones de relación en los modelos.
- list_relationships(): Enumera las relaciones entre tablas en un modelo.
- list_reports(): Devuelve una lista de informes.
- list_tables(): Enumera las tablas en un conjunto de datos.
- list_translations(): Proporciona las traducciones disponibles para los elementos en un modelo.
- list_workspaces(): Enumera los espacios de trabajo disponibles.

###### List Items = Lista de todos los objetos del área de trabajo

```python
#Devuelve una lista con los items de un área de trabajo especifica
#Parametros (type: Optional[str] = None, workspace: Union[str, uuid.UUID, NoneType] = None)) 
import sempy.fabric as fabric
x = fabric.list_items(workspace = area)
x
```

###### List Worskapce : Obtener lista de las Áreas de Trabajo

```python
#Devuelve todas las areas de trabajo que tenemos disponibles
#Parametros: (filter: Optional[str] = None, top: Optional[int] = None, skip: Optional[int] = None)##
    
import sempy.fabric as fabric
x = fabric.list_workspaces().sort_values(by='Name', ascending=True)
x
```
